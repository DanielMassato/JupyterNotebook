```python
import mlxtend
import pandas as pd 
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
produtos = pd.read_csv('bakery1000_dense.csv', delimiter=',')
freq_itemsets = apriori(produtos, min_support=0.01, use_colnames=True)
regras = association_rules(freq_itemsets, metric="lift", min_threshold=1)
regras.sort_values(by='lift',ascending=False).head(5)
```

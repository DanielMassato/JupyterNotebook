```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import parallel_coordinates


dados= pd.read_csv('iris.csv')
parallel_coordinates(dados, 'group')
plt.show()
```

```python
from sklearn import preprocessing
min_max_scaler=preprocessing.MinMaxScaler()
df_norm = min_max_scaler.fit_transform(dados)
dados_n = pd.DataFrame(df_norm,columns = dados.columns)
parallel_coordinates(dados_n,'group')
plt.show()
```

```python
#É possivel observar dois grupos pela diametro e largura da petala,  que se divide em duas regiões . 
```

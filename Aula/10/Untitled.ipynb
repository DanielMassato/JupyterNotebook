```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import StandardScaler
```

```python
centers = [[1, 1], [-1, -1], [1, -1]]
# gerando3 grupos artificiais centrados em “centers”
X, classes = make_blobs(n_samples=300, centers=centers,  cluster_std=0.4, random_state=0)
# normalizando os dados
X = StandardScaler ().fit_transform(X)
df = pd.DataFrame(X, columns=['x', 'y'])
plt.scatter(df['x'], df['y'])
plt.axis('off')
plt.show()
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=10)
rotulos= dbscan.fit_predict(df)
```

```python
plt.scatter(df['x'],df['y'], c=rotulos, cmap='rainbow') 
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
dbscan = DBSCAN(eps=0.1, min_samples=3)
rotulos= dbscan.fit_predict(df)
plt.scatter(df['x'],df['y'], c=rotulos, cmap='rainbow') 
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
df= pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.6, min_samples=5)
rotulos = dbscan.fit_predict(df)
```

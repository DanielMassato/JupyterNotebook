```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

pontos=[[5,3],[10,15],[15,12],[24,10],[30,45],[85,70],[71,80],[60,78],[55,52],[80,91]]
df = pd.DataFrame(pontos, columns=['x','y'])
plt.scatter(df['x'], df['y'])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
kmeans = KMeans(2)
rotulos = kmeans.fit_predict(df)
print(rotulos)
print(kmeans.cluster_centers_)
```

```python
plt.scatter(df['x'],df['y'], c=rotulos,cmap='rainbow')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
kmeans = KMeans(2)
rotulos = kmeans.fit_predict(df)
df_rep= pd.DataFrame(kmeans.cluster_centers_, columns=['x','y'])
plt.scatter(df['x'],df['y'], c=rotulos,cmap='rainbow')
plt.scatter(df_rep['x'], df_rep['y'],  cmap='rainbow')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

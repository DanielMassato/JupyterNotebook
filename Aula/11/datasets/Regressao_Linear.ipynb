#### Exemplo Scatterplot

```python
import matplotlib.pyplot as plt
import numpy as np
#cria a sequencia x = 0 a 1 em 5 intervalos (6 pontos)
x = np.linspace(0, 1,  6)
#cria a sequencia y = 0 a 5 em 5 intervalos (6 pontos)
y = np.linspace(0, 5,  6)
plt.scatter(x,y)
plt.show()
```

#### Regress√£o Linear Simples: Exemplo

```python
import pandas as pd
dados = pd.read_csv('lsd.csv ')

import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# DataFrame com os valores da vari√°vel independente
X = dados[['Tissue Concentration']]
# Vetor com os valores da vari√°vel dependente
y = dados[['Test Score' ]]
model = LinearRegression()
model.fit(X, y)

# valores reais na cor vermelha
plt.scatter(X, y,color='r')
# valores preditos em preto
plt.plot(X, model.predict(X),color='k')
# adicionando legenda aos eixos
plt.ylabel('Nota exame')
plt.xlabel('LSD conc.')
plt.show()
```

```python
model.predict(4)
```

#### Exerc√≠cio: Regressao Lineao do dataset Cars

```python
dados = pd.read_csv('cars.csv')
X = dados[['speed']]
# Vetor com os valores da vari√°vel dependente
y = dados[['dist']]
model = LinearRegression()
model.fit(X,y)
plt.scatter(X, y,color='r')
# valores preditos em preto
plt.plot(X, model.predict(X),color='k')
# adicionando legenda aos eixos
plt.ylabel('Distancia (pes)')
plt.xlabel('Velocidade (mph)')
plt.show()
```

```python
#qualidade do modelo criado: coeficiente R¬≤
model.score(X,y)
```

```python
#qual √© a predi√ß√£o para o valor de velocidade 30mph
model.predict(30)
```

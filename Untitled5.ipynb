```python
import mlxtend
import pandas as pd 
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
transacoes = pd.read_csv("compras_10.csv", delimiter=";")
transacoes = transacoes.set_index("transacao")
freq_itemsets = apriori(transacoes, min_support=0.1, use_colnames=True)
regras = association_rules(freq_itemsets, metric="lift", min_threshold=1)
```

```python
regras.sort_values(by='lift',ascending=False).head(5)
```
